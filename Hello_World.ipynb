```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
print(tf.__version__)
```

```python
# Build a simple Sequential model
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer = "sgd", loss = "mean_squared_error")
```

```python
# Declare model inputs and outputs for training
xs = np.array([-1,0,1,2,3,4], dtype = int)
ys = np.array([-3,-1,1,3,5,7], dtype = int)
```

**Training the Neural Network**:
The process of training the neural network, where it 'learns' the relationship between the x's and y's is in the model.fit() call. This is where it will go through the loop we spoke about above: making a guess, measuring how good or bad it is (aka the loss), using the optimizer to make another guess etc. It will do it for the number of epochs you specify. When you run this code, you'll see the loss on the right hand side.

```python
#The process of training the neural network, where it 'learns' the relationship between the inputs & outputs
model.fit(xs,ys,epochs =600)
```

**PREDICTION**:
Once the model has been trained to learn the relationship between x and y, you can use the model.predict() method to have it figure out a y for an unknown x.
So, for example, if x=10, what do you think y will be? Take a guess before you run this cod

```python
print(model.predict([10]))
```
